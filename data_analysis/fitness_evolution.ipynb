```python
# AVERAGE FITNESS PER GENERATION

import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data = pd.read_csv('../output/output.csv')

plt.figure(figsize=(10, 7))

avg_fitness_per_generation = data.groupby('Generation')['Fitness'].mean()
std_fitness_per_generation = data.groupby('Generation')['Fitness'].std()

plt.errorbar(avg_fitness_per_generation.index, avg_fitness_per_generation.values, 
             yerr=std_fitness_per_generation.values, marker='o', linestyle='-', 
             label='Average Fitness', color='b', capsize=5)

plt.xlabel('Generation')
plt.ylabel('Average Fitness')
plt.title('Average Fitness per Generation')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# AVERAGE FITNESS PER GENERATION FOR MULTIPLE GENERATIONS
# SAME AS THE ONE BEFORE BUT FOR TWO OR MORE EVOLUTIONS

import matplotlib.pyplot as plt
import pandas as pd

f1 = pd.read_csv('../output/output.csv')
f2 = pd.read_csv('../output/output1.csv')

# Load the data
data_list = [f1, f2]
colors = ['b', 'g']  
labels = ['Dataset 1', 'Dataset 2']  

plt.figure(figsize=(10, 7))

for i, data in enumerate(data_list):
    avg_fitness_per_generation = data.groupby('Generation')['Fitness'].mean()
    std_fitness_per_generation = data.groupby('Generation')['Fitness'].std()
    
    plt.errorbar(avg_fitness_per_generation.index, avg_fitness_per_generation.values, 
                 yerr=std_fitness_per_generation.values, marker='o', linestyle='-', 
                 label=labels[i], color=colors[i], capsize=5)

plt.xlabel('Generation')
plt.ylabel('Average Fitness')
plt.title('Average Fitness per Generation')
plt.legend()
plt.grid(True)

plt.show()
```
