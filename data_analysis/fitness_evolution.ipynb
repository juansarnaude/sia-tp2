```python
# AVERAGE FITNESS PER GENERATION

import matplotlib.pyplot as plt
import pandas as pd

# Load the data
data = pd.read_csv('../output/output.csv')

plt.figure(figsize=(10, 7))

avg_fitness_per_generation = data.groupby('Generation')['Fitness'].mean()
std_fitness_per_generation = data.groupby('Generation')['Fitness'].std()

plt.errorbar(avg_fitness_per_generation.index, avg_fitness_per_generation.values, 
             yerr=std_fitness_per_generation.values, marker='o', linestyle='-', 
             label='Average Fitness', color='b', capsize=5)

plt.xlabel('Generation')
plt.ylabel('Average Fitness')
plt.title('Average Fitness per Generation Archer')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# AVERAGE FITNESS PER GENERATION FOR MULTIPLE GENERATIONS
# SAME AS THE ONE BEFORE BUT FOR TWO OR MORE EVOLUTIONS

import matplotlib.pyplot as plt
import pandas as pd

f1 = pd.read_csv('../output_Archer/output_cutoff/cutoff_Optimum.csv')
f2 = pd.read_csv('../output_Mage/output_cutoff/cutoff_Content.csv')
f3 = pd.read_csv('../output_Warden/output_cutoff/cutoff_Structure.csv')
f4 = pd.read_csv('../output_Warrior/output_cutoff/cutoff_MaxGen.csv')



# Load the data
data_list = [f1, f2,f3,f4]
colors = ['b', 'g','y','r']  
labels = ['Archer', 'Mage', 'Warden', 'Warrior']  

plt.figure(figsize=(10, 7))

for i, data in enumerate(data_list):
    avg_fitness_per_generation = data.groupby('Generation')['Fitness'].mean()
    std_fitness_per_generation = data.groupby('Generation')['Fitness'].std()
    
    plt.errorbar(avg_fitness_per_generation.index, avg_fitness_per_generation.values, 
                 yerr=std_fitness_per_generation.values, marker='o', linestyle='-', 
                 label=labels[i], color=colors[i], capsize=5)

plt.xlabel('Generation')
plt.ylabel('Average Fitness')
plt.title('Average Fitness per Generation')
plt.legend()
plt.grid(True)

plt.show()
```

```python
#EVOLUTION OF ATTRIBUTES OF MOST FIT

import matplotlib.pyplot as plt
import pandas as pd



f1 = pd.read_csv('../output_Archer/output_cutoff/cutoff_Optimum.csv')
f2 = pd.read_csv('../output_Mage/output_cutoff/cutoff_Content.csv')
f3 = pd.read_csv('../output_Warden/output_cutoff/cutoff_Structure.csv')
f4 = pd.read_csv('../output_Warrior/output_cutoff/cutoff_MaxGen.csv')

f=[f1,f2,f3,f4]
names=["Archer","Mage","Warden","Warrior"]

for j,file in enumerate(f): 
    data_list = [file]
    colors = [ 'g', 'r', 'c', 'm', 'y', 'k']  
    attributes = [ 'Strength', 'Dexterity', 'Intelligence', 'Vigor', 'Constitution']
    
    plt.figure(figsize=(12, 8))
    color = 0
    
    for attr in attributes:
        
        for i, data in enumerate(data_list):
            fittest_per_generation = data.loc[data.groupby('Generation')['Fitness'].idxmax()]
    
            plt.plot(fittest_per_generation['Generation'], fittest_per_generation[attr], 
                     label=f'{attr}', color=colors[color], marker='o')
            color += 1
    
        
    plt.xlabel('Generation')
    plt.ylabel(f'Atributes Value')
    plt.title(f'Evolution of atributes for the Fittest Individual for {names[j]}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#EVOLUTION OF AVERAGE ATTRIBUTES


import matplotlib.pyplot as plt
import pandas as pd

f2 = pd.read_csv('../output/annular2.csv')

data_list = [f2]
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']  
attributes = ['Height', 'Strength', 'Dexterity', 'Intelligence', 'Vigor', 'Constitution']

plt.figure(figsize=(12, 8))
color = 0

for attr in attributes:
    for i, data in enumerate(data_list):
        avg_attr_per_generation = data.groupby('Generation')[attr].mean()
        std_attr_per_generation = data.groupby('Generation')[attr].std()

        plt.errorbar(avg_attr_per_generation.index, avg_attr_per_generation.values, 
                     yerr=std_attr_per_generation.values, label=f'{attr}', 
                     color=colors[color], marker='o', linestyle='-', capsize=5)

        color += 1  

plt.xlabel('Generation')
plt.ylabel('Attribute Value')
plt.title('Average Evolution of Attributes per Generation with Std Dev')
plt.legend()
plt.grid(True)

plt.show()
```
