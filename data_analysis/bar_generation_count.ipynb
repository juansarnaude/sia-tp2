```python
#AVERAGE GENERATION PER METHOD

import pandas as pd
import matplotlib.pyplot as plt

files_1 = ['../output/annular1.csv', '../output/annular2.csv', '../output/annular3.csv', '../output/annular4.csv', '../output/annular5.csv']
files_2 = ['../output/onepoint1.csv', '../output/onepoint2.csv', '../output/onepoint3.csv', '../output/onepoint4.csv', '../output/onepoint5.csv']
files_3 = ['../output/twopoint1.csv', '../output/twopoint2.csv', '../output/twopoint3.csv', '../output/twopoint4.csv', '../output/twopoint5.csv' ]
files_4 = ['../output/uniform1.csv', '../output/uniform2.csv', '../output/uniform3.csv', '../output/uniform4.csv', '../output/uniform5.csv']

tag_list = ['Anular', 'OnePoint', 'TwoPoint', 'Uniform']

generations_list = []

files = [files_1, files_2, files_3, files_4]

plt.figure(figsize=(10, 6))

for i, fileset in enumerate(files):
    for file in fileset:
        data = pd.read_csv(file)
        generations_list.append(data.iloc[-1]['Generation'])
        
    generation_series = pd.Series(generations_list)
    
    mean_generation = generation_series.mean()
    std_generation = generation_series.std()
    
    plt.bar([f'{tag_list[i]}'], [mean_generation], yerr=[std_generation], capsize=10, color='b', alpha=0.7)
    
    generations_list.clear()

plt.xlabel('Attributes')
plt.ylabel('Generation Number')
plt.title(f'Generations per crossover method')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
